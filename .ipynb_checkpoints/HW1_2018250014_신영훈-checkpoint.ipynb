```python
# Some basic setup. The libraries that you'll need to use will probably all be imported here.
# If you're unsure of what a certain library or tool does, Google it!

import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn import cluster
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale

%matplotlib inline
```

### (1) Implement K-means clustering algorithm and visualize data.
For given data, find the best number of clusters (each cluster is well-divided). <br>
Visualize your results using <b>scatter plot</b>. 

```python
# Use sklearn.cluster.KMeans for k-means clustering
# Use plt.scatter for visualization

n_samples = 3000
random_state = 156
data, _ = make_blobs(n_samples=n_samples, random_state=random_state)
```

```python
# Your code here. Define the variable values for X and y.
X = data[:,0]
y = data[:,1]
# End of code.

plt.scatter(X, y, s=5)
plt.title("Original Dataset Scatter Plot")
plt.show()
```

```python
# TODO: find the best number of clusters using your own method.
# You can either use a method introduced by Professor Kang during class or come up with your own.

# Your Code Here
distSqr = []
K = range(1,30)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(data)
    distSqr.append(kmeans.inertia_)
plt.plot(K, distSqr, 'rx-')
plt.xlabel('k')  
plt.ylabel('Sum of Squared Distance')
plt.show()

    
# End Your Code
```

## Explain your code.

Please explain the method you used in the cell above.
Write your answer in the cell below:

k값을 1~29까지 증가시켜보면서 각각의 경우 sample의 squared distance를 측정해 그려보았다. 
각각의 k값에 대해서 KMeans 모델을 훈련시킨 후, KMeans의 Attributes 중 sample들의 가장 가까운 cluster center까지의 거리의 제곱들의 합인 inertia_를 disSqr에 저장한 뒤, k에 대한 disSqr를 그래프로 그렸다. 
elbow는 k=3에서 나타났고, 따라서 k=3이 best number of clusters가 된다.

```python
# Your Code Here
BestClusterNumber = 3
# End Your Code

kmeans_model = KMeans(n_clusters=BestClusterNumber, random_state=1) 
kmeans_model.fit(data)

plt.scatter(X, y, c=kmeans_model.labels_, s=5)
plt.title("K-Means Result Scatter Plot")
plt.show()
```

### (2) Implement PCA and visualize data with class labels.
Conduct K-means clustering on given data. <br>
Implement <b>PCA(principle component analysis)</b> to convert high-dimensional vectors into 2-dimensional vectors. <br>
Compare plots by K-means result and class labels by visualization. <br>
* Dataset : Handwritten digit dataset (Class : digit, Data : digit image)
* Visualize <b>two scatter plots</b>. (One for class label and one for k-means clustering)

```python
# use sklearn.decomposition.PCA

digits = load_digits()
data = scale(digits.data)
labels = digits.target
```

```python
#TODO: Conduct K-means clustering on given data.
#TODO: Implement PCA to convert data into 2-d. 

# Your Code Here

# K-means
kmeans_digit = KMeans(n_clusters=10, random_state=1)
kmeans_digit.fit(data)

# PCA
pca = PCA(n_components=2)
new_data = pca.fit_transform(data)

# End Your Code
```

```python
# Visualize two scatter
# Your Code Here
X_pca = new_data[:, 0]
Y_pca = new_data[:, 1]
# End Your Code

colors = ['#e6194b', '#3cb44b', '#ffe119', '#4363d8', '#f58231', '#911eb4', '#46f0f0', '#f032e6', '#bcf60c', '#fabebe']

for i in range(10):
    px = X_pca[labels == i]
    py = Y_pca[labels == i]
    plt.scatter(px, py, c=colors[i], s=5, label=i)

plt.legend()    
plt.title("Class Label")
plt.show()

# Your Code Here
C_kmeans = kmeans_digit.labels_
# End Your Code

for i in range(10):
    px = X_pca[C_kmeans == i]
    py = Y_pca[C_kmeans == i]
    plt.scatter(px,py, c=colors[i], s=5, label=i)
    
plt.legend()
plt.title("K-means Clustering")
plt.show()
```
